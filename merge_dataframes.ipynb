# Analyse Umsatzzahlen Bäckereien in Kiel

```python
import pandas as pd


# Read the CSV files in seperate dataframes
df1 = pd.read_csv('umsatzdaten_gekuerzt.csv')
df2 = pd.read_csv('wetter.csv') 
df3 = pd.read_csv('kiwo.csv')

# Start with the sales data and merge all other data with a left-outer join
df = df1.merge(df2, on='Datum', how = 'left')
df = df.merge(df3, on='Datum', how = 'left')
```
